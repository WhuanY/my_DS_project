## 1 Order of EDA
* **Variable** - Variable name.
* **Type** - Identification of the variables' type. There are two possible values for this field: 'numerical' or 'categorical'. By 'numerical' we mean variables for which the values are numbers, and by 'categorical' we mean variables for which the values are categories.
* **Segment** - Identification of the variables' segment. We can define three possible segments: building, space or location. When we say 'building', we mean a variable that relates to the physical characteristics of the building (e.g. 'OverallQual'). When we say 'space', we mean a variable that reports space properties of the house (e.g. 'TotalBsmtSF'). Finally, when we say a 'location', we mean a variable that gives information about the place where the house is located (e.g. 'Neighborhood').
* **Expectation** - Our expectation about the variable influence in 'SalePrice'. We can use a categorical scale with 'High', 'Medium' and 'Low' as possible values.
* **Conclusion** - Our conclusions about the importance of the variable, after we give a quick look at the data. We can keep with the same categorical scale as in 'Expectation'.
* **Comments** - Any general comments that occured to us.

For **Expectation** we should think in the following ways:
* Do we think about this variable when we are buying a house? (e.g. When we think about the house of our dreams, do we care about its 'Masonry veneer type'?).
* If so, how important would this variable be? (e.g. What is the impact of having 'Excellent' material on the exterior instead of 'Poor'? And of having 'Excellent' instead of 'Good'?).
* Is this information already described in any other variable? (e.g. If 'LandContour' gives the flatness of the property, do we really need to know the 'LandSlope'?).

## 4.Missing Data
Important questions when thinking about missing data:

* How prevalent is the missing data?
* Is missing data random or does it have a pattern?
The answer to these questions is important for practical reasons because missing data can imply a reduction of the sample size. This can prevent us from proceeding with the analysis. Moreover, from a substantive perspective, we need to ensure that the missing data process is not biased and hidding an inconvenient truth.

```python
# Missing Data
# 得到所有属性的空值占比
total = df_train.isnull().sum().sort_values(ascending=False)
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False)
missing_data_percentage = pd.concat([total,percent],axis=1,keys=['total','percent'])
missing_data_percentage.head(20)
```

```python
df_train[['MasVnrArea','OverallQual']]
px.scatter(df_train[['MasVnrArea','OverallQual']],x='MasVnrArea',y='OverallQual')
```

让我们深入分析如何处理缺失数据。

我们将采用这样的标准：当数据缺失超过15%时，我们将删除相应的变量，就好像它们从未存在过一样。这意味着在这些情况下，我们不会尝试任何方法来填补缺失的数据。根据这个标准，有一组变量（例如 'PoolQC'、'MiscFeature'、'Alley' 等）应该被移除。关键问题是：我们会因为删除这些数据而有所遗漏吗？我认为不会。这些变量中没有一个看起来非常重要，因为在购房时，我们大多数时候并不会考虑它们（也许这就是数据缺失的原因之一？）。此外，更深入地研究这些变量，我们可以说 'PoolQC'、'MiscFeature' 和 'FireplaceQu' 这些变量很可能是异常值的强烈候选者，所以删除它们是明智之举。

至于其他情况，我们可以看到 'GarageX' 变量具有相同数量的缺失数据。我敢打赌，这些缺失数据可能来自同一组观测值（尽管我不会亲自验证，因为只有5%的缺失数据，不值得浪费太多时间来解决这个问题）。由于关于车库的最重要信息已经通过 'GarageCars' 表达出来，再加上我们只谈论5%的缺失数据，我决定删除这些 'GarageX' 变量。同样的逻辑也适用于 'BsmtX' 变量。

关于 'MasVnrArea' 和 'MasVnrType'，我们可以认为这些变量并不是十分关键。此外，它们与 'YearBuilt' 和 'OverallQual' 有很强的相关性，而这两个变量已经被考虑在内。因此，如果我们删除 'MasVnrArea' 和 'MasVnrType'，我们将不会丢失信息。

**最后，'Electrical' 变量中只有一个缺失观测值。由于只有一个观测值缺失，我们决定删除这个观测值，保留这个变量。**

总之，为了处理缺失数据，我们将删除所有具有缺失数据的变量，除了 'Electrical' 变量。在 'Electrical' 变量中，我们只会删除缺失数据的观测值。这样做可以确保我们的数据集更加清晰和可靠。

```python
#dealing with missing data
df_train = df_train.drop((missing_data_percentage[missing_data_percentage['total'] > 1]).index,1)
df_train = df_train.drop(df_train.loc[df_train['Electrical'].isnull()].index)
df_train.isnull().sum().max() #just checking that there's no missing data missing...
```

### Univariate Analysis
The primary concern here is to establish a threshold that defines an observation as an outlier. To do so, we'll standardize the data. In this context, data standardization means converting data values to have mean of 0 and a standard deviation of 1.

```python
# standardizing data
saleprice_scaled = StandardScaler().fit_transform(df_train[["SalePrice"]])
low_range = saleprice_scaled[saleprice_scaled[:,0].argsort()][:10]
high_range= saleprice_scaled[saleprice_scaled[:,0].argsort()][-10:]
print('outer range (low) of the distribution:')
print(low_range)
print('\nouter range (high) of the distribution:')
print(high_range)
```

How 'SalePrice' looks with her new clothes:

## Getting the hard core
们已经进行了一些数据清理，并了解了 'SalePrice' 的很多信息。现在是时候深入研究并了解 'SalePrice' 如何符合统计假设，以便我们能够应用多元分析技术。

根据Hair等人（2013）的观点，应该测试四个假设：

1. **正态性** - 当我们谈论正态性时，意味着数据应该呈现出正态分布的样子。这很重要，因为有些统计测试依赖于这一点（例如t-检验）。在这个练习中，我们将仅检查 'SalePrice' 的一元正态性（这是一个有限的方法）。请记住，一元正态性并不能确保多元正态性（这是我们希望拥有的），但它有所帮助。另一个需要注意的细节是，在大样本（>200个观测）中，正态性不是一个问题。然而，如果我们解决了正态性问题，我们可以避免许多其他问题（例如异方差性），这是我们进行这项分析的主要原因。

2. **同方差性** - 我希望我写得没错。同方差性是指 '假设依赖变量在预测变量的范围内展现出相等水平的方差'（Hair等人，2013）。同方差性是可取的，因为我们希望误差项在独立变量的所有值范围内都相同。

3. **线性性** - 评估线性性最常见的方法是检查散点图并寻找线性模式。如果模式不是线性的，那么值得探索数据转换。但是，我们不会深入探讨这个问题，因为我们看到的大多数散点图似乎具有线性关系。

4. **无相关误差** - 与定义一样，相关误差发生在一个误差与另一个误差之间存在相关性的情况下。例如，如果一个正误差系统地引发一个负误差，这意味着这些变量之间存在关系。这在时间序列中经常发生，因为某些模式与时间相关。我们也不会深入讨论这个问题。但是，如果你发现了某些情况，请尝试添加一个可以解释你获得的效应的变量。这是处理相关误差最常见的解决方案。



### 正态性检验
The point here is to test 'SalePrice' in a very lean way. We'll do this paying attention to:

- **Histogram** - Kurtosis and skewness.
- **Normal probability plot** - Data distribution should closely follow the diagonal that represents the normal distribution.

```python
## Histogram and Normal probability plot
sns.distplot(df_train['SalePrice'],fit=norm)
fig = plt.figure()
res = stats.probplot(df_train['SalePrice'],plot = plt)
print("Result of histplot and probPlot show that:\n the SalePrice may not derived from a normal distrubiton but with right skewness")
```

In summary, this code snippet is used for analyzing the 'SalePrice' data from df_train. 

1. It fisrt creates a histogram overlaid with a normal distribution curve to show the distribution of 'SalePrice'. 
2. Then, it creates a normal probability plot (or [Q-Q plot](https://library.virginia.edu/data/articles/understanding-q-q-plots)) to visually assess whether the 'SalePrice' data follows a normal distribution. 

This kind of analysis is common in statistical data analysis to understand the distribution characteristics of the data.

```python
#applying log transformation
df_train['SalePrice'] = np.log(df_train['SalePrice'])
```

Effectively, this replaces each value in the `"GrLivArea" `column with its <mark>natural logarithm</mark>. 

**Function:** This kind of transformation is often used in data preprocessing to <mark>help normalize data distributions or to reduce skewness</mark>, <u>making them more suitable for various statistical models and machine learning algorithms</u>.

```python
#transformed histogram and normal probability plot
sns.distplot(df_train['GrLivArea'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['GrLivArea'], plot=plt)
```

```python
#data transformation
df_train['GrLivArea'] = np.log(df_train['GrLivArea'])
df_train["GrLivArea"]
```

```python
#histogram and normal probability plot
sns.distplot(df_train['TotalBsmtSF'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train['TotalBsmtSF'], plot=plt)
```

Ok, now we are dealing with the big boss. What do we have here?

- **SalePrice**   Something that, in general, presents skewness.
- **TotalBsmtSF** A significant number of observations with value zero (houses without basement).
- **A big problem** because the value zero doesn't allow us to do log transformations.

To apply a log transformation here, we'll create a variable that can get the effect of having or not having basement (binary variable). Then, we'll do a log transformation to all the non-zero observations, ignoring those with value zero. This way we can transform data, without losing the effect of having or not basement.

I'm not sure if this approach is correct. It just seemed right to me. That's what I call 'high risk engineering'.

```python
# create column for new variable(one is enough beacuse)
df_train['HasBsmt'] = (df_train['TotalBsmtSF'] > 0).astype(int)
```

```python
df_train.loc[df_train["HasBsmt"]==1,'TotalBsmtSF'] = np.log(df_train["TotalBsmtSF"])
```

```python
#histogram and normal probability plot
sns.distplot(df_train[df_train['TotalBsmtSF']>0]['TotalBsmtSF'], fit=norm);
fig = plt.figure()
res = stats.probplot(df_train[df_train['TotalBsmtSF']>0]['TotalBsmtSF'], plot=plt)
```

### In the search for writing 'homoscedasticity(同方差性)' right at the first attempt

测试两个度量变量的**同方差性**的最佳方法是通过图形方式。不同于等方差性的情况会通过诸如锥形（图表一侧小离散度，另一侧大离散度）或菱形（分布中心有大量点）等形状来显示。

The best approach to test **homoscedasticity** for two metric variables is graphically. Departures from an equal dispersion are shown by such shapes as **cones** (small dispersion at one side of the graph, large dispersion at the opposite side) or **diamonds** (a large number of points at the center of the distribution).

```python
# Scatter Plot of SalePrice and Those who have Basement
plt.scatter(df_train[df_train['HasBsmt']==1]["TotalBsmtSF"]
           ,df_train[df_train['HasBsmt']==1]["SalePrice"])
```

Older versions of this scatter plot (previous to log transformations), had a conic shape (go back and check 'Scatter plots between 'SalePrice' and correlated variables (move like Jagger style)'). As you can see, the current scatter plot doesn't have a conic shape anymore. That's the power of normality! Just by ensuring normality in some variables, we solved the homoscedasticity problem.

Now let's check 'SalePrice' with 'TotalBsmtSF'.



```python
#scatter plot
plt.scatter(df_train[df_train['TotalBsmtSF']>0]['TotalBsmtSF'], df_train[df_train['TotalBsmtSF']>0]['SalePrice']);
```

We can say that, in general, 'SalePrice' exhibit equal levels of variance across the range of 'TotalBsmtSF'. Cool!
